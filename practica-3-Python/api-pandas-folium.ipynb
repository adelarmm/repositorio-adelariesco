En esta práctica, vamos a coger un conjunto de datos del ayuntamiento de Zaragoza y su apartado de tráfico con datos sobre los accidentes ocurridos en la ciudad. Vamos a usar un archivo csv que clasifica los datos según categorías (año, tipo, afectado, estado, vehículo), además de la calle en la que ha ocurrido el accidente.Esta ubicación geográfica nos resultará de interés para el análisis de los datos, por eso añadimos un mapa con las coordenadas de la ciudad.

Ahora le pedimos a Python que dentro de la librería folium, nos dé el folio Map, y seleccione las coordenadas que especificamos antes llamadas "geo_zrgz". Llamaremos a esta acción `mapa_zrgz`

```python
mapa_zrgz=folium.Map(location=geo_zrgz)
```

```python
mapa_zrgz
```

Vamos a leer los datos de la url, y para eso crearemos un dataframe como en la práctica de Api covid anterior. Por eso especificamos `df_zrgz`. Pedimos a panda que nos lea los datos `csv`, aunque no lo sean, por eso incluimos `delimiter`: le estamos aclarando a Python que de ese conjunto de datos, cada dato está delimitados por las comillas.

```python
df_zrgz=pd.read_csv(url_zrgz,delimiter=';')
```
